```python
#win32api library : a precision on the frequency of the sound and adjust its length to the milisecond.
#This opens the door for lots of sound creativity. 
#Frequency range:approx : 37 to 32000.
#Lower frequencies :avery deep sound , high frequencies:  high pitch sounds.
# random library’s ranint : a random integer between the values its given.
#datetime: for date and time,today for current date

import win32api
import random
import datetime
print (datetime.datetime.today().strftime("%H"),datetime.datetime.today().strftime("%M"))
hour=int(input("Enter a Hour: "))
minute=int(input("Enter a Minute: "))
while True:
    if hour == int(datetime.datetime.today().strftime("%H")) and minute == int(datetime.datetime.today().strftime("%M")):
        win32api.Beep(random.randint(42,1000), random.randint(450,3000))
        break
    else:
        print("Alarm Not Raised")
        break
```

```python
def find_Index(a):
    """ This functionn finds the type and index position of the input data
         a - input data for which the type and index has to be obtained"""
    if type(a)==str:
        print("The input is a string\n")
        for i in a:
            print("index of", i, 'is: ',a.index(i))
    elif type(a)==tuple:
        print("The input is a string\n")
        for i in a:
            print("index of", i, 'is: ',a.index(i)) 
    elif type(a)==set:
        print("The input is a string\n")
        a=list(a)
        for i in a:
            print("index of", i, 'is: ',a.index(i))           
    elif type(a)==list:
        print("The input is a string\n")
        for i in a:
            print("index of", i, 'is: ',a.index(i))
    elif type(a)==dict:
        print("The input is a  Dictionary\n")
        for i in a:
            a=list(a)
            print("index of key ", i, 'is: ',a.index(i))  
                  
    else:
        print('not an iterable')
```

```python
# Q16 : Write a fun which will take input as a dict and give me out as a list of all the values even in case of 2 level nesting it should work . 


```

```python
def func_dicttolist(d2):
    " " "This functionn takes dict and gives list eventhough dict within dict or list pressent in input data"""
    if type(d2) == dict:
        l=[]
        for i in d2:
            if type(d2[i])==int or type(d2[i])==str:
                l.append(d2[i])
            if type(d2[i])==list or type(d2[i])==tuple or type(d2[i])==set:
                l.extend(d2[i])
            if type(d2[i])==dict:
                v=d2[i].values()
                l.extend(v)
        return l 
    else:
        print(' not valid input as expecting  a dictionary')
```

```python
d2 = { 'key1' : "value"  , 234 : "xyz" , "name" : "sudh" , "no" : 34324345 ,'e':[2,3,4,6],'c':{'t','x','z'},'a':[1,6,8,'mi']}
func_dicttolist(d2)
```

```python
import PyPDF2
  
def PDF_merge(pdfs, output):
    """Merging the two pdfs function
    input: 2 pdf files"""
    # creating pdf file merger object
    pdfmerger = PyPDF2.PdfFileMerger()
  
    # appending pdfs one by one
    for pdf in pdfs:
        pdfmerger.append(pdf)
  
    # writing combined pdf to output pdf file
    with open(output, 'wb') as f:
        pdfmerger.write(f)

  
```

```python
# pdf files to merge
pdfs = ['D:\ineuron\linkin post\text.pdf', 'D:\ineuron\linkin post\voice.pdf']
    # output pdf file name
output = 'D:\ineuron\linkin post\combined_text_voice.pdf'
  
    # calling pdf merge function
PDF_merge(pdfs=pdfs, output=output)    
```
